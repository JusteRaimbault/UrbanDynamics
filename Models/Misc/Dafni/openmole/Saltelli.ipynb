```r
library(dplyr)
library(ggplot2)
library(reshape2)

firstOrder <- melt(as.tbl(read.csv('data/firstOrderIndices.csv')),id.vars = c('output'))
totalOrder <- melt(as.tbl(read.csv('data/totalOrderIndices.csv')),id.vars = c('output'))
data <- rbind(cbind(firstOrder,type=rep('First order',nrow(firstOrder))),cbind(totalOrder,type=rep('Total order',nrow(totalOrder))))
```

```r
options(repr.plot.width=10, repr.plot.height=5)
g=ggplot(data,aes(x=output,y=variable,fill=value))
g+geom_raster()+scale_fill_viridis_c()+facet_wrap(~type)
```
